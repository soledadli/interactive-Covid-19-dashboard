```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_complete  = pd.read_csv('Dataset_COVID_complete.csv')
```

```python
data_complete.shape
```

```python
data_complete.columns
```

```python
data_complete.head(3)
```

```python
data_complete.dtypes
```

```python
data_complete.groupby("Date")["Daily_Brazil_death"].sum().plot(figsize=(20,6))
plt.xlabel("Daily death", size=15)
plt.ylabel("Total cases", size=15)
plt.title("Evolution of daily death cases(Brazil)", size=15)
```

```python
data_complete.groupby(["month","year"])["Daily_Brazil_death"].sum().unstack().plot(figsize=(20,6))
plt.xlabel("Montly death", size=15)
plt.ylabel("Total cases", size=15)
plt.title("Evolution of death cases(Brazil)", size=15)
```

```python
data_complete.groupby(["month","year"])["Daily_India_death"].sum().unstack().plot(figsize=(20,6))
plt.xlabel("Montly death", size=15)
plt.ylabel("Total cases", size=15)
plt.title("Evolution of daily death cases(India)", size=15)
```

```python
data_complete.plot(x="Date", y=["Daily_Brazil_confirmed", "Daily_France_confirmed", "Daily_US_confirmed",
                                "Daily_India_confirmed","Daily_South_Africa_confirmed"],
                  figsize=(16, 16))
plt.xlabel("Daily confirmed cases", size=15)
plt.ylabel("Total cases", size=15)
plt.title("Comparative of daily confirmed cases", loc='left', fontsize=12, fontweight=0, color='orange')
plt.show()
```

```python
data_complete.plot(x="Date", y=["Daily_Brazil_recovered", "Daily_France_recovered", "Daily_US_recovered",
                                "Daily_India_recovered","Daily_South_Africa_recovered"],
                  figsize=(16, 16))
plt.xlabel("Daily recovered cases", size=15)
plt.ylabel("Total cases", size=15)
plt.title("Comparative of daily recovered cases", loc='center', fontsize=12, fontweight=2, color='brown')
plt.show()
```

```python
data_complete.plot(x="Date", y=["Daily_Brazil_death", "Daily_France_death", "Daily_US_death",
                                "Daily_India_death","Daily_South_Africa_death"],
                  figsize=(16, 12))
plt.xlabel("Daily death cases", size=15)
plt.ylabel("Total cases", size=15)
plt.title("Comparative of daily death cases", loc='right', fontsize=12, fontweight=4, color='aqua')
plt.show()
```
